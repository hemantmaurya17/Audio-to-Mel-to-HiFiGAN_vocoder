```python
import os
import json
import numpy as np
import librosa
import torch
from types import SimpleNamespace
from models import Generator
from scipy.io.wavfile import write
import matplotlib.pyplot as plt
import librosa.display
from IPython.display import Audio, display

# Project directory set karo jahan model files hain
os.chdir("/home/hemant/Desktop/my_project/hifi-gan")  # apna path lagao
```

```python
# Config load karo
with open("config_v1.json") as f:
    config_dict = json.load(f)
config = SimpleNamespace(**config_dict)

# Generator load karo
model = Generator(config)
checkpoint = torch.load("generator_v1.pt", map_location="cpu")
model.load_state_dict(checkpoint["generator"])
model.eval()
print("✅ HiFi-GAN Model Loaded Successfully")
```

```python
def wav_to_stft_and_mel(file_path, config):
    # Audio load
    y, sr = librosa.load(file_path, sr=config.sampling_rate)
    y = y / np.abs(y).max() * 0.99  # normalize

    # STFT
    stft = librosa.stft(y, n_fft=config.n_fft, hop_length=config.hop_size, win_length=config.win_size)
    stft_db = librosa.amplitude_to_db(np.abs(stft), ref=np.max)

    plt.figure(figsize=(10, 4))
    librosa.display.specshow(stft_db, sr=sr, hop_length=config.hop_size, x_axis='time', y_axis='hz')
    plt.colorbar(format="%+2.0f dB")
    plt.title("STFT Spectrogram")
    plt.tight_layout()
    plt.show()

    # Mel Spectrogram
    mel = librosa.feature.melspectrogram(
        y=y,
        sr=sr,
        n_fft=config.n_fft,
        hop_length=config.hop_size,
        win_length=config.win_size,
        n_mels=config.num_mels,
        fmin=config.fmin,
        fmax=config.fmax
    )
    mel_db = librosa.power_to_db(mel, ref=np.max)

    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mel_db, sr=sr, hop_length=config.hop_size, x_axis='time', y_axis='mel')
    plt.colorbar(format="%+2.0f dB")
    plt.title("Mel Spectrogram")
    plt.tight_layout()
    plt.show()

    mel_tensor = torch.from_numpy(mel).unsqueeze(0).float()
    return mel_tensor
```

```python
def mel_to_audio(mel_tensor, model, config, output_file="output_from_scale.wav"):
    with torch.no_grad():
        audio_out = model(mel_tensor).squeeze().cpu().numpy()

    # Save karo
    write(output_file, config.sampling_rate, (audio_out * 32767).astype(np.int16))
    print(f"✅ Audio Generated Successfully: {output_file}")

    # Waveform plot
    plt.figure(figsize=(10, 3))
    plt.plot(audio_out)
    plt.title("Reconstructed Waveform")
    plt.xlabel("Samples")
    plt.ylabel("Amplitude")
    plt.tight_layout()
    plt.show()

    # Playback for quality check
    display(Audio(output_file))
    return audio_out
```

```python
input_wav = "scale.wav"   # apna audio file
mel = wav_to_stft_and_mel(input_wav, config)
audio = mel_to_audio(mel, model, config, "output_from_scale.wav")
```
