```python

%matplotlib inline
import torch
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
import json
from types import SimpleNamespace
import soundfile as sf

# Quality metrics
from pesq import pesq
from pystoi import stoi
import python_speech_features as psf
from dtw import accelerated_dtw


```

```python
# 1. Load Audio

AUDIO_PATH = "lj.wav"  
sr = 22050

y, sr = librosa.load(AUDIO_PATH, sr=sr)
print(f"âœ… Loaded audio: {y.shape}, sr={sr}")

plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Original Waveform")
plt.show()
ipd.Audio(y, rate=sr)
```

```python
# 2. HiFi-GAN Preprocessing (official)

from meldataset import mel_spectrogram   # HiFi-GAN repo se import hona chahiye

# config.json se match hone chahiye
n_fft = 1024
num_mels = 80
sampling_rate = 22050
hop_size = 256
win_size = 1024
fmin = 0
fmax = 8000

audio_torch = torch.FloatTensor(y).unsqueeze(0)  # (1, T)
mel = mel_spectrogram(
    audio_torch, n_fft, num_mels, sampling_rate,
    hop_size, win_size, fmin, fmax
)

print("âœ… Mel shape:", mel.shape)
```

```python
from models import Generator  # HiFi-GAN repo ke models.py

CONFIG_PATH = "config_v1.json"
CHECKPOINT_PATH = "generator_v1.pt"

with open(CONFIG_PATH) as f:
    config = json.load(f)
config = SimpleNamespace(**config)

model = Generator(config)
checkpoint = torch.load(CHECKPOINT_PATH, map_location="cpu")
model.load_state_dict(checkpoint["generator"])
model.eval()
print("âœ… HiFi-GAN model loaded")
```

```python
# 4. HiFi-GAN Inference
with torch.no_grad():
    y_hat = model(mel).squeeze().cpu().numpy()

sf.write("hifigan_output.wav", y_hat, sr)
print("âœ… HiFi-GAN reconstructed audio saved")
ipd.Audio(y_hat, rate=sr)
```

```python
# 5. Griffin-Lim Reconstruction
mel_librosa = librosa.feature.melspectrogram(
    y=y, sr=sr, n_fft=n_fft, hop_length=hop_size,
    win_length=win_size, n_mels=num_mels, fmin=fmin, fmax=fmax
)
mel_inv = librosa.feature.inverse.mel_to_stft(mel_librosa, sr=sr, n_fft=n_fft)

y_gl = librosa.griffinlim(
    mel_inv, n_iter=60, hop_length=hop_size, win_length=win_size
)

sf.write("griffinlim_recon.wav", y_gl, sr)
print("âœ… Griffin-Lim reconstructed audio saved")
ipd.Audio(y_gl, rate=sr)
```

```python
# 6. Compare Waveforms
plt.figure(figsize=(12, 6))
plt.subplot(3, 1, 1)
librosa.display.waveshow(y, sr=sr)
plt.title("Original Audio")

plt.subplot(3, 1, 2)
librosa.display.waveshow(y_hat, sr=sr)
plt.title("HiFi-GAN Reconstructed")

plt.subplot(3, 1, 3)
librosa.display.waveshow(y_gl, sr=sr)
plt.title("Griffin-Lim Reconstructed")

plt.tight_layout()
plt.show()
```

```python
# 7. Quality Metrics
def evaluate_metrics(ref, deg, sr=16000):
    ref, _ = librosa.load(ref, sr=sr)
    deg, _ = librosa.load(deg, sr=sr)

    min_len = min(len(ref), len(deg))
    ref, deg = ref[:min_len], deg[:min_len]

    # PESQ
    pesq_score = pesq(sr, ref, deg, "wb")
    # STOI
    stoi_score = stoi(ref, deg, sr, extended=False)
    # MCD
    ref_mfcc = psf.mfcc(ref, samplerate=sr, numcep=13)
    deg_mfcc = psf.mfcc(deg, samplerate=sr, numcep=13)
    _, cost, _, _ = accelerated_dtw(ref_mfcc, deg_mfcc, dist='euclidean')
    mcd_score = float(np.mean(cost)) / len(ref_mfcc)
    return float(pesq_score), float(stoi_score), float(mcd_score)

print("\nðŸ“Š Quality Metrics Comparison")
gl_pesq, gl_stoi, gl_mcd = evaluate_metrics(AUDIO_PATH, "griffinlim_recon.wav")
hg_pesq, hg_stoi, hg_mcd = evaluate_metrics(AUDIO_PATH, "hifigan_output.wav")

print(f"Griffin-Lim -> PESQ: {gl_pesq:.3f}, STOI: {gl_stoi:.3f}, MCD: {gl_mcd:.3f} dB")
print(f"HiFi-GAN    -> PESQ: {hg_pesq:.3f}, STOI: {hg_stoi:.3f}, MCD: {hg_mcd:.3f} dB")
```
